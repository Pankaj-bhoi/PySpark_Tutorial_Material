```
# Read Student.csv file
# Create a new col in the DF for total marks and let the total marks be 120
# Create new column avg to calculate the avg marks of the student (marks/total marks)*100
# filter out all those students who have achieved more than 80% marks in OOP course and save it in new DF
# filter out all those students who have achieved more than 60% marks in cloud course and save it in new DF
# print the names and marks of all the students from the above DFs.
```

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Read files').getOrCreate()
df = spark.read.option("header",True).csv('/FileStore/tables/StudentData.csv')
df.show()
```

```
df.printSchema()
```

```
from pyspark.sql.functions import col,lit
df1 = df.withColumn("Total Marks",lit(120))
df1.show()
```

```
df2 = df1.withColumn('Avg Marks',(col('marks')/col('Total Marks'))*100)
df2.show()
```

```
df_OOP = df2.filter((df2.course == 'OOP') & (col('Avg Marks') > 80))
df_OOP.show()
```

```
df_Cloud = df2.filter((df2.course == 'Cloud') & (col('Avg Marks')>60))
df_Cloud.show()
```

```
df_OOP.select("name","marks").show()
df_Cloud.select('name','marks').show()
```
