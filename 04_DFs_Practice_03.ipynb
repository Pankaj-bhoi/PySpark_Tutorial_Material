```
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/StudentData.csv')
df_sum = df.groupBy("gender").sum('marks')
df_count = df.groupBy("course").count()
df_max = df.groupBy("gender").max("marks")
df_min = df.groupBy("gender").min('marks')
df_mean = df.groupBy("gender").mean('marks')
df_avg = df.groupBy('gender').avg('age')
print('SUM')
df_sum.show()
print('COUNT')
df_sum.show()
print('MAX')
df_max.show()
print('MIN')
df_mean.show()
print('MEAN')
df_mean.show()
print('AVG')
df_avg.show()
```

```
# groupBy( ) in multiple columns and aggregation
from pyspark.sql.functions import sum,count,min,max,mean,avg
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/StudentData.csv')
df = df.groupby('gender').agg(count('name').alias('Total_name'),sum('marks').alias('SUM'),min('marks').alias('MIN'),max('marks').alias('MAX'),\
                             mean('marks').alias('MEAN'),avg('marks').alias('AVG')).show()
```

```
from pyspark.sql.functions import col
df = spark.read.options(header=True, inferSchema=True).csv('/FileStore/tables/StudentData.csv')

# Transformation
df2 = df.filter(df.gender=='Male')
df3 = df2.groupby('course', 'gender').agg(count("*").alias('Total_enroll'))
df3.filter(df3.Total_enroll > 85).show()
# Alternate
df.filter(df.gender=='Male').groupby('course','gender').agg(count('*').alias('Total_enroll')).filter(col('Total_enroll') >85).show() # Use 'col' for "Total_enroll"
```
