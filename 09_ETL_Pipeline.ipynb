```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('ETL job').getOrCreate()
df = spark.read.text('/FileStore/tables/WordData.txt')
df.show()
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit, col, explode
import pyspark.sql.functions as f

df2 = df.withColumn('Splited Data', f.split('value', ' '))
df3 = df2.withColumn('words',explode('Splited Data')).select('words').groupby('words').count()
df3.show()
```

```
driver = 'org.postgresDriver'
url = "endpoint-----point----of---url"
table = 'database_name.table_name'
user = 'your username'
password = "your password"

df3.write.format('jdbc').option("driver",driver).option("url",url).option("table",table).option("user",user).option("password",password).save()
```
