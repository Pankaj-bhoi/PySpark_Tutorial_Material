```
# Mini Project
#Student files
rdd = sc.textFile('/FileStore/tables/StudentData.csv')
header = rdd.first()
rdd1 = rdd.filter(lambda x: x != header)
rdd1.count() # No of Student
```

```
# Total marks achieved by male and female
rdd2 = rdd1.map(lambda x: x.split(','))
rdd3 = rdd2.map(lambda x:(x[1], int(x[5])))
rdd3.reduceByKey(lambda x,y: x+y).collect()
```

```
# Total number of students have passed and failed
passed = rdd2.filter(lambda x: int(x[5])>50).count()
failed = rdd2.filter(lambda x: int(x[5])<50).count()
print(f'Passed : {passed} | failed : {failed}')
```

```
# Total number of students enrolled per course
rdd2.map(lambda x: (x[3], 1 )).reduceByKey(lambda x,y: x+y).collect()
```

```
# Total marks that achieved by students per course
rdd2.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x,y:x+y).collect()
```

```
# Average marks that students have achieved per course
rdd2.map(lambda x: (x[3], (int(x[5]),1))).reduceByKey(lambda x,y:(x[0]+y[0], x[1]+y[1])).map(lambda x:x[1][0]/x[1][1]).collect()
# Alternate 
rdd2.map(lambda x: (x[3], (int(x[5]),1))).reduceByKey(lambda x,y:(x[0]+y[0], x[1]+y[1])).mapValues(lambda x: x[0]/x[1]).collect()
```

```
# Minimum and Maximum marks achieved by students per course
minimum = rdd2.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x,y: x if x<y else y).collect()
maximum = rdd2.map(lambda x: (x[3], int(x[5]))).reduceByKey(lambda x,y: x if x>y else y).collect()
print(f"minimum : {minimum}")
print(f'maximum : {maximum}')
```
