```
from pyspark import SparkContext, SparkConf
from pyspark.streaming import StreamingContext

conf = SparkConf().setAppName('Steam data')
sc = SparkContext.getOrCreate(conf=conf)
ssc = StreamingContext(sc,1)

# Alternate option
# ssc_Alternate = StreamingContext.getActiveOrCreate(sc,1)

rdd = ssc.textFileStream('/FileStore/tables')
rdd = rdd.map(lambda x: (x,1))
rdd = rdd.reduceByKey(lambda x,y: x+y)
rdd.pprint()
ssc.start()
ssc.awaitTerminationOrTimeout(100)
```

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('DF streaming').getOrCreate()
word = spark.readStream.text('/FileStore/tables')
word.writeStream.format("console").outputMode('complete').start()
```

```
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('Df stream').getOrCreate()
df = spark.readStream.text('/FileStore/tables/stream')
df = df.groupBy("value").count()
display(df)
```
